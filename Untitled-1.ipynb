```python
#Buat Skrip Pelatihan Model

train_df = pd.read_csv("train_dataset.csv")
test_df = pd.read_csv("test_dataset.csv")

# Rapikan nama kolom
train_df.columns = train_df.columns.str.strip().str.lower()
test_df.columns = test_df.columns.str.strip().str.lower()


# ENCODE LABEL
train_df["label_numeric"] = train_df["label"].map({"GON+": 1, "GON-": 0})
test_df["label_numeric"] = test_df["label"].map({"GON+": 1, "GON-": 0})


# PREPARE IMAGE PATH
image_folder = "images_resized"

# LOAD TRAINING DATA
X_train, y_train = [], []

for _, row in train_df.iterrows():
    img_path = os.path.join(image_folder, row["image name"])

    if os.path.exists(img_path):
        try:
            img = Image.open(img_path).convert("RGB").resize((224, 224))
            img_array = np.array(img) / 255.0

            X_train.append(img_array)
            y_train.append(row["label_numeric"])
        except:
            continue

X_train = np.array(X_train)
y_train = np.array(y_train)

print("Train shape:", X_train.shape)


# LOAD TEST DATA
X_test, y_test = [], []

for _, row in test_df.iterrows():
    img_path = os.path.join(image_folder, row["image name"])

    if os.path.exists(img_path):
        try:
            img = Image.open(img_path).convert("RGB").resize((224, 224))
            img_array = np.array(img) / 255.0

            X_test.append(img_array)
            y_test.append(row["label_numeric"])
        except:
            continue

X_test = np.array(X_test)
y_test = np.array(y_test)

print("Test shape:", X_test.shape)


# BUILD CNN MODEL
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(224,224,3)),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()


# TRAIN MODEL
history = model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.2
)


# EVALUATE
test_loss, test_acc = model.evaluate(X_test, y_test)
print("Test Accuracy:", test_acc)

# SAVE MODEL
model.save("glaucoma_model.h5")
print("Model saved!")
```

```python
# LOAD MODEL
model = tf.keras.models.load_model("glaucoma_model.h5")

# LOAD DATASET
test_df = pd.read_csv("test_dataset.csv")

# Rapikan kolom
test_df.columns = test_df.columns.str.strip().str.lower()

# Encode label
test_df["label_numeric"] = test_df["label"].map({"GON+": 1, "GON-": 0})

# LOAD IMAGE DATA
image_folder = "images_resized"

X_test, y_test = [], []

for _, row in test_df.iterrows():
    img_path = os.path.join(image_folder, row["image name"])

    if os.path.exists(img_path):
        try:
            img = Image.open(img_path).convert("RGB").resize((224, 224))
            img_array = np.array(img) / 255.0

            X_test.append(img_array)
            y_test.append(row["label_numeric"])
        except:
            continue

X_test = np.array(X_test)
y_test = np.array(y_test)

print("Test data shape:", X_test.shape)

# PREDICTION
predictions = model.predict(X_test)

# Convert ke 0 / 1
predicted_labels = (predictions > 0.5).astype(int).reshape(-1)

# CONFUSION MATRIX
cm = confusion_matrix(y_test, predicted_labels)

print("Confusion Matrix:")
print(cm)

np.save("confusion_matrix.npy", cm)

# CLASSIFICATION REPORT
print("\nClassification Report:")
print(classification_report(y_test, predicted_labels))

report = classification_report(y_test, predicted_labels, output_dict=True)
accuracy = report["accuracy"]
precision = report["1"]["precision"]   
recall = report["1"]["recall"]
f1 = report["1"]["f1-score"]
metrics = {
    "accuracy": accuracy,
    "precision": precision,
    "recall": recall,
    "f1_score": f1
}
with open("metrics.json", "w") as f:
    json.dump(metrics, f)

# VISUALISASI CONFUSION MATRIX
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Non-Glaucoma", "Glaucoma"],
            yticklabels=["Non-Glaucoma", "Glaucoma"])

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
#visualisasi model prediksi

# LOAD MODEL
model = tf.keras.models.load_model("glaucoma_model.h5")

# LOAD DATASET
test_df = pd.read_csv("test_dataset.csv")
test_df.columns = test_df.columns.str.strip().str.lower()

# bikin label numeric
test_df["label_numeric"] = test_df["label"].map({"GON+": 1, "GON-": 0})

image_folder = "images_resized"

# PREPARE DATA
X_test, y_test, valid_image_names = [], [], []

for _, row in test_df.iterrows():
    image_path = os.path.join(image_folder, row["image name"])

    if os.path.exists(image_path):
        try:
            img = Image.open(image_path).convert("RGB")
            img = img.resize((224, 224))
            img_array = np.array(img) / 255.0

            X_test.append(img_array)
            y_test.append(row["label_numeric"])
            valid_image_names.append(row["image name"])
        except:
            continue
    else:
        print(f"Missing image: {image_path}")

X_test = np.array(X_test)
y_test = np.array(y_test)

print("Total test images loaded:", len(X_test))

# PREDICT
predictions = model.predict(X_test)
predicted_labels = (predictions > 0.5).astype(int)

# LABEL MAP
label_map = {1: "Glaucoma", 0: "Normal"}

# INTERPRETASI PROB
def interpret_probability(prob):
    if prob >= 0.95:
        return "Very confident glaucoma"
    elif prob >= 0.70:
        return "Moderate confidence"
    elif prob >= 0.52:
        return "Uncertain prediction"
    else:
        return "Low probability of glaucoma"

# DISPLAY 10 DATA
print("\nDisplaying first 10 predictions...")

for i in range(min(10, len(X_test))):
    image_path = os.path.join(image_folder, valid_image_names[i])
    img = Image.open(image_path)

    actual = y_test[i]
    predicted = predicted_labels[i][0]
    probability = predictions[i][0]

    plt.figure(figsize=(4,4))
    plt.imshow(img)
    plt.title(
        f"Actual: {label_map[actual]} | Predicted: {label_map[predicted]}\n"
        f"Prob: {probability:.2f} | {interpret_probability(probability)}"
    )
    plt.axis("off")
    plt.show()

# ERROR ANALYSIS
incorrect_indices = np.where(predicted_labels.flatten() != y_test)[0]
print("Number of incorrect predictions:", len(incorrect_indices))

# TAMPILKAN SALAH
for idx in incorrect_indices[:5]:
    image_path = os.path.join(image_folder, valid_image_names[idx])
    img = Image.open(image_path)

    actual = y_test[idx]
    predicted = predicted_labels[idx][0]
    probability = predictions[idx][0]

    plt.figure(figsize=(4,4))
    plt.imshow(img)
    plt.title(
        f"Actual: {label_map[actual]} | Predicted: {label_map[predicted]}\n"
        f"Prob: {probability:.2f} | {interpret_probability(probability)}"
    )
    plt.axis("off")
    plt.show()

# RANDOM SAMPLE
print("\nRandom predictions...")

for idx in random.sample(range(len(X_test)), min(5, len(X_test))):
    image_path = os.path.join(image_folder, valid_image_names[idx])
    img = Image.open(image_path)

    actual = y_test[idx]
    predicted = predicted_labels[idx][0]
    probability = predictions[idx][0]

    plt.figure(figsize=(4,4))
    plt.imshow(img)
    plt.title(
        f"Actual: {label_map[actual]} | Predicted: {label_map[predicted]}\n"
        f"Prob: {probability:.2f}"
    )
    plt.axis("off")
    plt.show()
```
